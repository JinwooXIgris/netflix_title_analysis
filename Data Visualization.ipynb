## Data Visualization

- Visual representation of data to discover trends and patterns
- Helps in storytelling and decision-making

## Types

- Categorical: Bar chart, Pie chart, etc
- Numerical: Box plot, Histogram, etc
- Relationship: Scatter plot, Line plot, Heatmap, Pair plot, etc

## Libraries

- Matplotlib
- Seaborn

```python
import pandas as pd

import matplotlib.pyplot as plt
```

- figure: The entire canvas that holds all subplots(A figure can hold 1 or more plots)
- Axes: A single plot

```python
#create a figure with a single axes

x = [1, 2, 3]

y = [4, 5, 6]

#fig and ax are conventional ways to name it but not compulsory

fig, ax = plt.subplots()
ax.plot(x, y)

#Title

ax.set_title("Basic Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

#Always call the plt.show  to show your plot

plt.show()
```

```python
# create a figure with multiple subplots
#subplots come in rows and columns
#If you want to 2 plots side by side: one row two columns

fig, axes = plt.subplots(1, 2)
axes[0].plot(x, y)
axes[1].plot(x, y)

#Title
axes[0].set_title("Left Plot")
axes[1].set_title("Right Plot")

plt.show()
```

```python
fig, axes = plt.subplots(2, 2)

axes[0, 0].plot(x, y)
axes[0, 1].plot(x, y)
axes[1, 0].plot(x, y)
axes[1, 1].plot(x, y)

plt.show()
```

## 21st Class

 What are the number of netflixtitles released per year?

```python
title_per_year = df.groupby("release_year")["title"].count().reset_index()

title_per_year
```

```python
# Since the 2 columns are numerical We'll use a line plot
#Line plot
#You can add a color

plt.plot(title_per_year["release_year"], title_per_year["title"], color = 'red', marker = "o")
plt.title("Number of Netflix Titles Released Per year")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")

plt.show()
```

### Bivariate Analysis

- Examining the relationship between two variables

```python
#Movies and TV shows released per year

release_year_by_type = df.groupby(["release_year", "type"]).size().unstack(fill_value = 0).reset_index()  #Size will count the number of observations for each year for eachcategories
                                                                              #Unstack to make it a dataframe
release_year_by_type
```

```python
df.groupby(["release_year", "type"]).size()
```

```python
#Unstack to make it a dataframe

df.groupby(["release_year", "type"]).size().unstack(fill_value = 0).reset_index()
```

```python
plt.plot(release_year_by_type["release_year"], release_year_by_type["Movie"], label = "Movies")
plt.plot(release_year_by_type["release_year"], release_year_by_type["TV Show"], label = "Tv Shows")
plt.legend()

plt.show()
```
